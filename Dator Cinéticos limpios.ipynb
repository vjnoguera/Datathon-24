```python
import pandas as pd
import numpy as np
```

```python
df_cin_cen = pd.read_csv("Datos Cinéticos\Cinéticos IPC centrifugacion.csv", sep = ";", encoding="ANSI")
df_cin_cen
```

```python
df_cin_cen = df_cin_cen.iloc[:,:5]
df_cin_cen.info()
```

```python
df_cin_cen["Fecha"] = pd.to_datetime(df_cin_cen["Fecha"])
```

```python
df_cin_cen = df_cin_cen.applymap(lambda x: x.replace(',', '.') if isinstance(x, str) else x)
df_cin_cen[df_cin_cen.select_dtypes(include=['object']).columns] = df_cin_cen.select_dtypes(include=['object']).apply(lambda col: pd.to_numeric(col, errors='coerce'))
```

```python
df_cin_cen.info()
```
